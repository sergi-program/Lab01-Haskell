## Pregunta teórica

Haskell es un lenguaje **fuertemente tipado** porque cada expresión tiene un tipo definido y no permite mezclar tipos incompatibles.

También es **evaluado perezosamente** porque las expresiones solo se evalúan cuando su resultado es necesario.

## Definición de funciones

```haskell
-- doble de un numero
doble :: Int -> Int
doble x = x * 2
doble 7
```

```haskell
-- funcion que determina si es mayor de edad
esMayorEdad :: Int -> Bool
esMayorEdad edad = edad >= 18

esMayorEdad 20
esMayorEdad 15
```

## Diferencia entre Int y Float

- **Int** representa números enteros (sin decimales).
- **Float** representa números decimales.
